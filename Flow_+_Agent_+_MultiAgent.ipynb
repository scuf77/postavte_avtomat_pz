```python
import os
from ddgs import DDGS
from agno.agent import Agent

os.environ["OPENAI_API_KEY"] = "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"  # Groq key
os.environ["OPENAI_BASE_URL"] = "https://api.groq.com/openai/v1"
```

```python
def web_search(query: str) -> list[str]:
    with DDGS() as ddg:
        results = ddg.text(query, max_results=3)
        return [r["body"] for r in results]

def extract_name(texts: list[str]) -> str:
    joined = " ".join(texts).lower()
    if "дарья" in joined:
        return "Дарья Зотеева"
    return "Не удалось определить"

def flow_real_name():
    texts = web_search("реальное имя Инстасамка")
    name = extract_name(texts)
    print(name)

print("Flow")
flow_real_name()
```

```python
def get_search_snippets(query: str) -> str:
    with DDGS() as ddg:
        results = ddg.text(query, max_results=3)
        return "\n".join(r["body"] for r in results)

agent = Agent(
    name="SingleAgent",
    instructions=(
        "You are given web search snippets. "
        "State the real full name of Инстасамка. "
        "Answer with the name only."
    ),
    model="openai:llama-3.3-70b-versatile",
    markdown=False
)

search_data = get_search_snippets("реальное имя Инстасамка")

print("Agent")
result = agent.run(search_data)
print(result.content)
```

```python
collector = Agent(
    name="Collector",
    instructions=(
        "From the given web snippets, list all possible real names "
        "associated with Инстасамка."
    ),
    model="openai:llama-3.3-70b-versatile",
    markdown=False
)

verifier = Agent(
    name="Verifier",
    instructions=(
        "You are given candidate names. "
        "Choose the correct real full name of Инстасамка. "
        "Answer with one name only."
    ),
    model="openai:llama-3.3-70b-versatile",
    markdown=False
)

print("Team")

search_data = get_search_snippets("реальное имя Инстасамка")

candidates = collector.run(search_data).content
final_answer = verifier.run(candidates).content

print("TEAM RESULT:", final_answer)
```

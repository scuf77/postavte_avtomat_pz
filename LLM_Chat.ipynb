```python
import os
from browser_use_sdk import BrowserUse
from rich import print
from rich.panel import Panel
from rich.console import Console
from rich.theme import Theme

from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("BROWSER_USE_API_KEY")

if not api_key:
    print("[red]Ошибка: Не удалось загрузить API ключ[/red]")
    print("[cyan]Проверьте файл .env в текущей папке[/cyan]")
    exit()

console = Console(theme=Theme({
    "header": "#b38dd9",
    "border": "#f3efe8"
}))

try:
    client = BrowserUse(api_key=api_key)
    print("[green]Подключение к API успешно[/green]")
except Exception as e:
    print(f"[red]Ошибка подключения: {e}[/red]")
    exit()

chat_history = [
    {
        "role": "system",
        "content": "Я AI-помощник. Отвечаю на вопросы подробно и понятно."
    }
]

console.print(
    Panel(
        "Здравствуйте! Я AI-ассистент. Здавайте вопрос",
        title="[border]AI Помощник[/border]",
        border_style="border"
    )
)

while True:
    question = input("Ваш вопрос: ")

    if question.lower() in ["exit", "выход"]:
        console.print(
            Panel(
                "Завершаю работу. До новых встреч!",
                title="[red]Завершение[/red]",
                border_style="red"
            )
        )
        break

    elif question.lower() in ["clear", "очистить"]:
        chat_history = [chat_history[0]]
        console.print("[green]История диалога очищена[/green]")

    elif question.lower() in ["system", "настройка"]:
        console.print("[green]Введите новую инструкцию для бота:[/green]")
        new_instruction = input()
        if new_instruction:
            chat_history[0]["content"] = new_instruction
            console.print("[green]Настройки обновлены[/green]")

    elif question.lower() in ["help", "помощь"]:
        console.print(Panel(
            "[cyan]Доступные команды:[/cyan]\n"
            "[green]выход[/green] - завершить работу\n"
            "[green]очистить[/green] - очистить историю\n"
            "[green]настройка[/green] - изменить системное сообщение\n"
            "[green]помощь[/green] - показать это сообщение",
            title="[border]Помощь[/border]",
            border_style="border"
        ))

    elif question.lower() in ["web", "веб"]:
        console.print("[cyan]Введите задание для поиска в интернете:[/cyan]")
        web_task = input("Задание: ")
        try:
            task = client.tasks.create_task(
                task=web_task,
                llm="browser-use-llm"
            )
            result = task.complete()

            console.print(
                Panel(
                    result.output,
                    title="[border]Результат поиска[/border]",
                    border_style="border"
                )
            )
        except Exception as error:
            console.print(f"[red]Ошибка: {error}[/red]")

    else:
        try:
            task = client.tasks.create_task(
                task=f"Ответь на вопрос: {question}",
                llm="browser-use-llm"
            )
            result = task.complete()

            answer = result.output

            console.print(
                Panel(
                    answer,
                    title="[border]Ответ[/border]",
                    border_style="border"
                )
            )

        except Exception as error:
            console.print(f"[red]Ошибка: {error}[/red]")
            console.print("[yellow]Возможные причины:[/yellow]")
            console.print("[yellow]- Неправильный API ключ[/yellow]")
            console.print("[yellow]- Закончились токены на аккаунте[/yellow]")
            console.print("[yellow]- Проблемы с подключением к серверу[/yellow]")
```
